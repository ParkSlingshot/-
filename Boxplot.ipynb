```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
plt.rc('font', family='NanumBarunGothic')
```

#3. Load & Select Data

```python
file = '/gdrive/MyDrive/data/data_2020.csv'
data = pd.read_csv(file)
```

#4. Histogram

```python
plt.figure(figsize=(20,10))
sns.histplot(data = data[data['도착영업소코드']==110]['통행시간'], color = 'green', label = '목천')
sns.histplot(data = data[data['도착영업소코드']==125]['통행시간'], color = 'blue', label = '남구미')
sns.histplot(data = data[data['도착영업소코드']==140]['통행시간'], color = 'red', label = '부산')
plt.legend(title = '도착영업소')
plt.show()
```

#5. Box plot

```python
plt.figure(figsize=(20,10))
sns.set(style = 'ticks', palette='pastel')

sns.boxplot(x='도착영업소코드', y='통행시간',data = data)
plt.show()
```

```python
#extraordinary 값을 나타냄
data_stat = data[data['도착영업소코드']==130]['통행시간'].describe()
data_stat
```

```python
#요일별 boxplot
plt.figure(figsize=(20,10))
sns.set(style = 'ticks', palette= 'pastel')

sns.boxplot(x='요일' , y = '통행시간', data=data)
plt.show()
```

```python
#집계시당 통행시간
x=data['집계시']
x=[str(i)+'시' for i in x]



plt.figure(figsize=(20,10))
sns.set(style='ticks', palette= 'pastel')

sns.boxplot(x= x,  y='통행시간', data= data)
plt.legend(title='집계시')
plt.show()
```
