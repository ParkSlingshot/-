#1. Linear function

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
Hare_speed = 2 #@param{type : "slider", min : 0, max : 10, step:1}

MAXVAL =10
INTERVAL =MAXVAL + 1
h_xdata, h_ydata = [], []
```

```python
plt.figure (figsize = (5,5))
for t in np.linspace(0,MAXVAL, INTERVAL):
  h_y = Hare_speed * t
  h_xdata.append(t)
  h_ydata.append(h_y)

plt.plot(h_xdata, h_ydata, 'ro', label="Hare")

plt.title('Linear Regression', fontsize = 16)
plt.xlabel('time')
plt.ylabel('distance')
plt.legend()
plt.show()
```

#2. Hypothesis

```python
velocity_Variance = 0.2 #@param {type:"slider", min:0, max:2, step:0.1}
Lines = 5

a_val= Hare_speed + (velocity_Variance * Lines)
h_xdata, h_ydata, v_xdata, v_ydata = [],[],[],[]

a_val
```

```python
plt.figure(figsize = (5,5))

list = np.linspace(0,MAXVAL,INTERVAL)

for t in list: 
  h_y = Hare_speed * t
  h_xdata.append(t)
  h_ydata.append(h_y)
  a = a_val - (t * velocity_Variance)
  for i in list:
    h_y = a*i
    v_xdata.append(i)
    v_ydata.append(h_y)
  plt.plot(v_xdata, v_ydata, alpha=0.2)

plt.plot(h_xdata, h_ydata, 'ro', label="Hare")
plt.title('Linear Regression')
plt.xlabel('Time')
plt.ylabel('Distance')
plt.legend()
plt.show()
```

#3. Calculate Cost (gradient descent)

```python
import matplotlib.animation as anm
from matplotlib import rc
rc('animation', html = 'jshtml')

#Counfigue
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111)
ax.set_xlim(-0.1,2.1)
ax.set_ylim(0,400)
t_xdata, t_ydata = [],[]

#animation

def get_cost(a_val):
  cost = 0
  for i in list:
    cost += pow((a_val*i - Hare_speed*i) ,2) 
  return cost

def animateFrame(frame):
  a_val = Hare_speed + (velocity_Variance * Lines)
  i = frame * velocity_Variance
  a = a_val - i
  t_xdata.append(i )
  t_ydata.append(get_cost(a))
  plot = ax.plot(t_xdata, t_ydata, 'ro')
  return plot

ani = anm.FuncAnimation(fig, animateFrame, frames=list, blit=True, repeat = False  )

ax.set_title('Gradient Descent')
ax.set_ylabel('Total Cost')
ax.set_xlabel('Varience')

ani
```
