```python
image = cv2.imread('cat.jpg')
```

```python
print("width : {} pixels \n height : {} pixels \n channel : {} pixles".format(image.shape[1], image.shape[0], image.shape[2]))
```

```python
x = 132 #@param{type:"slider", min:0, max:275, step:1}
y= 0 #@param{type:"slider", min:0, max:183, step:1}
size = 168 #@param{type:"slider", min:0, max:200, step:1}

width,height = image.shape[:2]
center = (width//2, height//2)
```

#OpenCV Transform

```python
cv2_imshow(image)
```

```python
#Move the image
move = np.float32([[1,0,-50],[0,1,-50]])
moved = cv2.warpAffine(image, move, (width, height))
cv2_imshow(moved)
```

```python
#Rotate the image
rotate = cv2.getRotationMatrix2D(center,90,1)
rotated = cv2.warpAffine(image, rotate, (width, height))
cv2_imshow(rotated)
```

```python
#scailing
def rescailing (frame, scale = 0.55):
  width = int(frame.shape[1]*scale)
  height = int(frame.shape[0]*scale)
  dimension = (width, height)
  return cv2.resize(frame, dimension, interpolation = cv2.INTER_AREA)

resized = rescailing(image)
cv2_imshow(resized)
```

```python
#Flip
flipped = cv2.flip(image,-1)
cv2_imshow(flipped)
```
