```python
image = cv.imread('cat.jpg')
def rescale (frame, scale = 1.5):
  width = image.shape[1] * scale
  height = image.shape[0] * scale
  dimension = (width, height)
  return cv.resize(image, dimension, interpolation=cv.INTER_AREA)
```

```python
gray = cv.cvtColor(image, cv.COLOR_RGB2GRAY)
cv_imshow(gray)
```

#Edge Detection (Laplacian, HSV. Threshold, Contour)

```python
#Laplacian
lap = cv.Laplacian(gray, cv.CV_64F)
cv_imshow(lap)
```

```python
#HSV
hsv = cv.cvtColor(image, cv.COLOR_BGR2HSV)
cv_imshow(hsv)
```

```python
#Threshold
ret,thresh = cv.threshold(gray, 150,75,0)
cv_imshow(thresh)
```

```python
#Contours
contours, hirechy = cv.findContours(thresh, cv.RETR_LIST, cv.CHAIN_APPROX_SIMPLE)
contours = sorted(contours,key=cv.contourArea, reverse=True)
for i in range(len(contours)):
  cv.drawContours(image,contours, i ,(150,75,0))
cv_imshow(image)
```
