#1. Linear Function
Tortoise & Hare

```python
#import library
import numpy as np
import math
import matplotlib.pyplot as plt

#Paremeters
Tortoise_Speed = 2 #@param{type : "slider", min:0, max:10, step: 1}
Tortoise_Bias = 5 #@param{type : "slider", min:0, max:10, step:1}
Hare_Speed = 3 #@param{type: "slider", min:0, max:10, step:1}
t_xdata, t_ydata, h_xdata, h_ydata = [],[], [], []
doMeet = False

#time 
for t in np.linspace(0,10,101) :
  t_y = Tortoise_Speed * t + Tortoise_Bias
  h_y = Hare_Speed * t
  t_xdata.append(t)
  t_ydata.append(t_y)
  h_xdata.append(t)
  h_ydata.append(h_y)
  if h_y >= t_y and (not(doMeet)):
    doMeet =True
    meetTime = t
    meetDistance = t_y
plt.figure(figsize = (10,10))
plt.plot(t_xdata, t_ydata, label ="Tortoise")
plt.plot(h_xdata, h_ydata, label ="Hare")
plt.xlabel('km',fontsize =13)
plt.ylabel('Hour',fontsize =13)
plt.legend
if (doMeet) : 
  plt.title('Meet ' + str(math.ceil(meetTime))+'hour ' + str(math.ceil(meetDistance))+'km')
```
