#Face Detection with haar_cascade

```python
obama = cv.imread('obama.jpg')
gray_obama = cv.cvtColor(obama, cv.COLOR_RGB2GRAY)
```

```python
players = cv.imread('soccer.jpg')

def rescale(frame, scale=2):
  width = frame.shape[1]*scale
  height = frame.shape[0]*scale
  dimension = (width, height)
  return cv.resize(frame, dimension, interpolation = cv.INTER_AREA)

player = rescale(players)

gray_player = cv.cvtColor(player, cv.COLOR_BGR2GRAY)
```

```python
face_scade = cv.CascadeClassifier('front_face.xml')
face_rect = face_scade.detectMultiScale(gray_player, scaleFactor=1.1, minNeighbors=6)

for (x,y,w,h) in face_rect:
  cv.rectangle(player, (x,y),(x+w,y+h), (0,255,255), 2)
```

```python
face_scade = cv.CascadeClassifier('front_face.xml')
face_rect = face_scade.detectMultiScale(gray_obama, scaleFactor=1.1, minNeighbors=7)

for (x,y,w,h) in face_rect:
  cv.rectangle(obama, (x,y),(x+w,y+h), (0,255,255), 2)
```
